Here we are looking at the run that Ryan added fixes to to prevent leaves growing when the status is zero. As we will see, this killed off most of the northern latitudes.

He here implement a fix that stops carbon stores going to zero on this line
https://github.com/NGEET/fates/blob/393b905468d93786735a340356d48babc3769988/biogeochem/EDPhysiologyMod.F90#L940

and thus avoids the subsequent termination mortality which was killing of high latitude plants (that were getting intot his space by virtue of not having enough in their store).
